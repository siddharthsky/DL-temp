```python
# Resolving the meta data problem xD
df[["video_id","title", "channel_name", "views", "description", "datePublished", "duration"]]=0
def get_video_info(url):
    global tt,cn,vie,des,dt_pub,dur,tgs,idg
    tt,cn,vie,des,dt_pub,dur,tgs,idg = "","","","","","","",""
    
    response = urlopen(url)
    soup = bs(response, "html.parser")
    
    idg = url.split("=")[-1]
    tt = soup.find("meta", itemprop="name")['content']
    cn = soup.find("span", itemprop="author").next.next['content']
    vie = soup.find("meta", itemprop="interactionCount")['content']
    des = soup.find("meta", itemprop="description")['content']
    dt_pub = soup.find("meta", itemprop="datePublished")['content']
    dur = soup.find("span", {"class": "ytp-time-duration"})
    tgs = ', '.join([ meta.attrs.get("content") for meta in soup.find_all("meta", {"property": "og:video:tag"}) ])
    return tt,cn,vie,des,dt_pub,dur,tgs,idg

for idx,col in enumerate(video_urls):
    dd = df["Links"][idx]
    print(dd)
    get_video_info(dd)
    df.loc[idx, "video_id"] = idg
    df.loc[idx, "title"] = tt
    df.loc[idx, "channel_name"] = cn
    df.loc[idx, "views"] = vie
    df.loc[idx, "description"] = des
    df.loc[idx, "datePublished"] = dt_pub
    df.loc[idx, "duration"] = dur
    #df["tags"][idx] = tgs
```
