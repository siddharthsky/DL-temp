```python
import pandas as pd
import os
import re
from requests_html import HTMLSession 
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs # importing BeautifulSoup
```

```python
df= pd.read_excel("NASA2.xlsx")
```

```python
df["Links"]
```

```python
df[["title", "channel_name", "views", "description", "datePublished", "duration", "tags"]]=0
```

```python
def get_video_info(url):
    global tt,cn,vie,des,dt_pub,dur,tgs
    tt,cn,vie,des,dt_pub,dur,tgs = "","","","","","",""
    
    response = urlopen(url)
    soup = bs(response, "html.parser")
    
    tt = soup.find("meta", itemprop="name")['content']
    cn = soup.find("span", itemprop="author").next.next['content']
    vie = soup.find("meta", itemprop="interactionCount")['content']
    des = soup.find("meta", itemprop="description")['content']
    dt_pub = soup.find("meta", itemprop="datePublished")['content']
    dur = soup.find("span", {"class": "ytp-time-duration"})
    tgs = ', '.join([ meta.attrs.get("content") for meta in soup.find_all("meta", {"property": "og:video:tag"}) ])
    return tt,cn,vie,des,dt_pub,dur,tgs
```

```python
for idx,col in enumerate(df["Links"]):
    dd = df["Links"][idx]
    print(dd)
    get_video_info(dd)
    df["title"][idx] = tt
    df["channel_name"][idx] = cn
    df["views"][idx] = vie
    df["description"][idx] = des
    df["datePublished"][idx] = dt_pub
    df["duration"][idx] = dur
    df["tags"][idx] = tgs
```
